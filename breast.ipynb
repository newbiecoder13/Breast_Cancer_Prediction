```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("./breast-cancer.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.iloc[:,1]=le.fit_transform(df.iloc[:,1].values)
```

```python
df.tail(15)
```

```python
X=df.iloc[:,2:]
Y=df.iloc[:,1]
```

```python
X.head(50)
```

```python
Y.head(50)
```

```python
from sklearn.model_selection import train_test_split as tts
X_train,X_test,Y_train,Y_test=tts(X,Y,test_size=0.2,random_state=21)
```

```python
from sklearn.preprocessing import StandardScaler
ss= StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.fit_transform(X_test)
```

```python
X_train
```

```python
import tensorflow as tf
```

```python
model=tf.keras.models.Sequential([
    tf.keras.layers.Dense(128,activation='relu',input_dim=30),
    tf.keras.layers.Dense(128,activation='relu'),
    tf.keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',metrics=['accuracy'])
```

```python
hist=model.fit(X_train,Y_train,batch_size=100,epochs=50)
```

```python
predict=model.predict(X_test)
predict=(predict>0.5)
```

```python
predict
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,predict)
sns.heatmap(cm,annot=True,cmap='YlGnBu')
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy: ",accuracy_score(Y_test,predict)*100)
```
